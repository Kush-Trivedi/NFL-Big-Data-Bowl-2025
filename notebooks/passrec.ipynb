```python
def print_passer_receiver_summary(df):
    pass_result_mapping = {
        'C': 'Completed',
        'I': 'Incomplete',
        'S': 'Short',
        'IN': 'Interception',
        'R': 'Rejected'
    }
    df['Pass Result'] = df['passResult'].map(pass_result_mapping)
    df['Pass Result'] = df['Pass Result'].fillna(df['passResult'])
    aggregation_functions = {
        'pass_count': 'sum',
        'C %': 'mean',
        'I %': 'mean'
    }
    grouped = df.groupby(['passer_player_name', 'receiver_player_name']).agg(aggregation_functions).reset_index()
    grouped = grouped.rename(columns={
        'pass_count': 'Total Passes',
        'C %': 'C Percentage',
        'I %': 'I Percentage'
    })
    completed = df[df['passResult'] == 'C'].groupby(['passer_player_name', 'receiver_player_name'])['pass_count'].sum().reset_index()
    completed = completed.rename(columns={'pass_count': 'Total Completed'})
    summary = pd.merge(grouped, completed, on=['passer_player_name', 'receiver_player_name'], how='left')
    summary['Total Completed'] = summary['Total Completed'].fillna(0)
    summary['Completion Rate (%)'] = (summary['Total Completed'] / summary['Total Passes']) * 100
    summary['Completion Rate (%)'] = summary['Completion Rate (%)'].round(2)
    preferred_locations = df.groupby(['passer_player_name', 'receiver_player_name', 'pass_location'])['pass_count'].sum().reset_index()
    preferred = preferred_locations.loc[preferred_locations.groupby(['passer_player_name', 'receiver_player_name'])['pass_count'].idxmax()]
    preferred = preferred.rename(columns={'pass_location': 'Preferred Location', 'pass_count': 'Preferred Location Passes'})
    preferred = pd.merge(preferred, grouped[['passer_player_name', 'receiver_player_name', 'Total Passes']], 
                         on=['passer_player_name', 'receiver_player_name'], how='left')
    preferred['Preferred Location %'] = (preferred['Preferred Location Passes'] / preferred['Total Passes']) * 100
    preferred['Preferred Location %'] = preferred['Preferred Location %'].round(2)
    summary = pd.merge(summary, preferred[['passer_player_name', 'receiver_player_name', 'Preferred Location', 'Preferred Location %']], 
                       on=['passer_player_name', 'receiver_player_name'], how='left')
    summary = summary.sort_values(by='Total Passes', ascending=False)

    for _, row in summary.iterrows():
        passer = row['passer_player_name']
        receiver = row['receiver_player_name']
        total_passes = row['Total Passes']
        c_percent = row['C Percentage']
        i_percent = row['I Percentage']
        preferred_location = row['Preferred Location']
        preferred_location_percent = row['Preferred Location %']
        
        print(f"{passer} to {receiver}:")
        print(f"  Total passes: {total_passes}")
        print(f"  C %: {c_percent:.2f}%")
        print(f"  I %: {i_percent:.2f}%")
        print(f"  Preferred location: {preferred_location} side ({preferred_location_percent:.2f}% of passes)")
        print("-" * 50)


def print_player_summary(df):
    df['route %'] = pd.to_numeric(df['route %'], errors='coerce').fillna(0)
    grouped = df.groupby('displayName')
    summary_list = []
    for name, group in grouped:
        routes_sorted = group.sort_values(by='route %', ascending=False)
        route_list = [f"{row['routeRan']}: {row['route %']}" for _, row in routes_sorted.iterrows()]
        routes_str = "%, ".join(route_list)
        summary_list.append({
            'displayName': name,
            'Routes Ran': routes_str
        })
    summary_df = pd.DataFrame(summary_list)
    for _, row in summary_df.iterrows():
        name = row['displayName']
        routes = row['Routes Ran']
        
        print(f"Player Name: {name}")
        print(f"Routes Ran: {routes}")
        print("-" * 50)

ari_pass_rec_df = pd.read_csv("assets/offesnse-data/ARI/ARI_pass_receiver_analysis.csv")
print_passer_receiver_summary(ari_pass_rec_df)
ari_route = pd.read_csv("assets/offesnse-data/ARI/ARI_route_analysis.csv")
print_player_summary(ari_route)
```

```python
import re
import pandas as pd

def name_match(display_name, receiver_name):
    # Check if either name is NaN
    if pd.isna(display_name) or pd.isna(receiver_name):
        return False
    
    # Split names into parts
    display_parts = display_name.split()
    receiver_parts = receiver_name.split('.')
    
    # Check if the first initial matches
    display_initial = display_parts[0][0].lower()  # First initial in displayName
    receiver_initial = receiver_parts[0][0].lower()  # First initial in receiver_name
    
    # Combine remaining parts and normalize by removing punctuation and converting to lowercase
    display_last = re.sub(r'[^a-zA-Z]', '', ''.join(display_parts[1:])).lower()
    receiver_last = re.sub(r'[^a-zA-Z]', '', ''.join(receiver_parts[1:])).lower()
    
    return display_initial == receiver_initial and display_last == receiver_last

def print_passer_receiver_summary(df_pass, df_route):
    summary = df_pass[["passer_player_name", "receiver_player_name", "total_passes", "C %", "I %", "C right %", "C left %", "C middle %"]].copy()
    summary = summary[summary['total_passes'] > 1]
    summary['max_pct'] = summary[['C right %', 'C left %', 'C middle %']].max(axis=1)
    summary['preferred_side'] = summary[['C right %', 'C left %', 'C middle %']].idxmax(axis=1).str.replace('C ', '').str.replace(' %', '')
    summary['preferred_location'] = summary['max_pct'].astype(str) + '% on ' + summary['preferred_side']
    summary = summary.sort_values('max_pct', ascending=False)
    summary = summary.drop_duplicates(subset=['passer_player_name', 'receiver_player_name'], keep='first')
    summary = summary.drop(['max_pct', 'preferred_side',"C right %", "C left %", "C middle %"], axis=1)
    summary = summary.sort_values(by='total_passes', ascending=False)
    summary = summary.reset_index(drop=True)
    for _, row in summary.iterrows():
        passer = row['passer_player_name']
        receiver = row['receiver_player_name']
        total_passes = row['total_passes']
        c_percent = row['C %']
        i_percent = row['I %']
        preferred_location = row['preferred_location']
        matched_display_name = None

        for display_name in df_route['displayName'].unique():
            if name_match(display_name, receiver):
                matched_display_name = display_name
                break

        if matched_display_name:
            receiver_routes = df_route[df_route['displayName'] == matched_display_name]
            routes_sorted = receiver_routes.sort_values(by='route %', ascending=False)
            route_list = [f"{r['routeRan']}: {r['route %']}%" for _, r in routes_sorted.iterrows()]
            routes_str = "\n    - ".join(route_list)
        else:
            routes_str = "N/A"
        
        print(f"Player Name: {passer} to {receiver}")
        print(f"  Total Passes: {total_passes}")
        print(f"  C %: {c_percent:.2f}%")
        print(f"  I %: {i_percent:.2f}%")
        print(f"  Routes Ran:\n    - {routes_str}")
        print(f"  Preferred location: {preferred_location}")
        print("-" * 50)
```
